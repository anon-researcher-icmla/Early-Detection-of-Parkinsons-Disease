```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Gait_Data.csv')

df
```

```python
# Dropping rows with NA/Null values

df.dropna(inplace=True)

df
```

```python
# Drop non-numeric and identifier columns
df = df.drop(columns=['PATNO', 'EVENT_ID', 'INFODT'], errors='ignore')
df.shape
```

```python
# Identify the target variable (assuming COHORT as the target for classification)
target = 'COHORT'
X = df.drop(columns=[target], errors='ignore')
y = df[target]
```

```python
# Correlation Analysis: Remove highly correlated features (threshold > 0.9)
corr_matrix = X.corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
highly_correlated = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
X_corr_filtered = X.drop(columns=highly_correlated, errors='ignore')
```

```python
# Mutual Information: Rank top 10 features
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(X_corr_filtered, y, discrete_features=False, random_state=42)
mi_feature_importance = pd.Series(mi_scores, index=X_corr_filtered.columns).sort_values(ascending=False).head(10)
```

```python
# Random Forest Feature Importance: Rank top 10 features
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_corr_filtered, y)
rf_feature_importance = pd.Series(rf.feature_importances_, index=X_corr_filtered.columns).sort_values(ascending=False)
```

```python
# Display results
feature_selection_results = pd.DataFrame({
    "Mutual Information Score": mi_feature_importance,
    "Random Forest Importance": rf_feature_importance
}).fillna(0)

print("Feature Selection Results:")
print(feature_selection_results)
```
